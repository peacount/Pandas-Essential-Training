```python
lo1 = oo[oo.NOC == 'USA']
lo2 = lo1.groupby(['Athlete', 'Discipline']).size()
lo2.sort_values(ascending=False).head()
```

Alternative Solution

```python
gy = oo[oo.NOC == 'USA']
gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
```

```python
gy.reset_index(inplace=True)

for year, group in gy.groupby('Edition'):
    print(group.sort_values('Total', ascending=False))
```

```python
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')]
tu
```

```python
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
